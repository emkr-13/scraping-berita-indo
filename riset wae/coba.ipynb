```python
import mechanicalsoup

# Create a Browser object
browser = mechanicalsoup.Browser()

# Specify the URL of the webpage you want to scrape
url = "https://www.google.com/search?q=ganjar+::+tribunnews&sca_esv=729edc120b86c4d9&tbs=cdr:1,cd_min:1/15/2024,cd_max:1/15/2024&tbm=nws&sxsrf=ACQVn098j05T6oxrzqg5oovlwz5QSvu38g:1706603595333&ei=S7S4ZYHzE7uR4-EPq5CB2Ak&start=70&sa=N&ved=2ahUKEwjBxMno2YSEAxW7yDgGHStIAJsQ8tMDegQIBBAQ&biw=1366&bih=654&dpr=1"

# Use the Browser to open the URL
response = browser.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Print the HTML content of the page
    print(response.text)
else:
    print(f"Failed to fetch the webpage. Status code: {response.status_code}")
```
